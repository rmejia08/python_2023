```python
# Muestre estas medidas de tendencia central en un BoxPlot para cada variable mencionada.
rs = data

plt.figure(figsize=(8, 6))
# CREAMOS FUNCION PARA
def boxplot_bd(data,column):
    sns.boxplot(
         data=rs,
         x=column,
         color='Blue'
    )

    plt.axvline(# Línea horizontal para la media
        x=rs[column].mean(),
        color='red',
        linestyle='--',
        label='Media'
    )

    plt.show()
    print("\n")

boxplot_bd(rs, 'MORA60')
boxplot_bd(rs, 'ATRASO')
boxplot_bd(rs, 'EDAD')
boxplot_bd(rs, 'TC')
boxplot_bd(rs, 'AHORRO')
boxplot_bd(rs, 'INGRESO')
boxplot_bd(rs, 'LINEA_TC')
boxplot_bd(rs, 'SALDO_DEUDOR')
boxplot_bd(rs, 'SCORE')
boxplot_bd(rs, 'ABONO_PAGO')
boxplot_bd(rs, 'RIESGO_SBS')
```

NIVEL INTERMEDIO

```python
# ¿Hay alguna variable con valores faltantes en el conjunto de datos?

df = data.isnull().melt()

sns.displot(data=df,
            y='variable',
            hue='value',
            multiple='fill',
            aspect=2)
plt.show()
```

```python
# Genere un gráfico de barras para todas las variables numéricas.

rs = data

def cotplot(data,column):
  sns.catplot(
      data=rs,
      x=column,
      kind='count'
  )
  plt.show()

cotplot(rs, 'MORA60')
cotplot(rs, 'ATRASO')
cotplot(rs, 'EDAD')
cotplot(rs, 'TC')
cotplot(rs, 'AHORRO')
cotplot(rs, 'INGRESO')
cotplot(rs, 'LINEA_TC')
cotplot(rs, 'SALDO_DEUDOR')
cotplot(rs, 'SCORE')
cotplot(rs, 'ABONO_PAGO')
cotplot(rs, 'RIESGO_SBS')
```

```python
#VARIABLES CATEGORICAS
d1 = data['TIPOCASA'].value_counts(sort=True).reset_index(name='count')
d2 = data['DEPARTAMENTO'].value_counts(sort=True).reset_index(name='count')
d3 = data['EDUCACION'].value_counts(sort=True).reset_index(name='count')


sns.barplot(data=d1,
            x='index',
            y='count')
plt.show()

sns.barplot(data=d2,
            x='index',
            y='count')
plt.xticks(rotation=90)
plt.show()

sns.barplot(data=d3,
            x='index',
            y='count')
plt.xticks(rotation=90)
plt.show()
```

```python
# Grafique la correlación entre INGRESO y el monto de la línea de crédito de tarjeta de crédito.

sns.displot(
    data=data,
    x='INGRESO',
    y='LINEA_TC'
)
plt.show()
```

```python
# Realice un diagrama de dispersión entre la EDAD y el RIESGO_SBS.
sns.scatterplot(x='EDAD', y='RIESGO_SBS', data=data)

plt.xlabel('Edad')
plt.ylabel('Riesgo SBS')
plt.title('Diagrama de dispersión entre EDAD y RIESGO_SBS')

plt.show()
```

```python
# Realice un headmap para TIPOCASA y EDUCACION
COR = pd.crosstab(data['TIPOCASA'],data['EDUCACION'])
sns.heatmap(COR)
plt.show()
```

```python
#Realice un headmap para EDAD y RIESGO_SBS.
COR = pd.crosstab(data['EDAD'],data['RIESGO_SBS'])
sns.heatmap(COR)
plt.show()
```

NIVEL AVANZADO

```python
# ¿Qué porcentaje de clientes ha tenido una morosidad mayor a 60 días (`MORA60`) después de la aprobación del crédito?

moro = data[data['ATRASO'] > 60]

porcen = (len(moro) / len(data)) * 100

print(f"Porcentaje de clientes con morosidad mayor a 60 días: {porcen:.2f}%")
```

```python
# Investigue si existe alguna relación entre el nivel educativo y la incidencia de morosidad mayor a 60 días. ¿Cómo podría visualizar esta relación?

tab = pd.crosstab(data['MORA60'],data['EDUCACION'])

sns.heatmap(tab)
plt.show()
```
